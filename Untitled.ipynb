```python
import pandas as pd
import seaborn as sns
import numpy as np
import glob
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib as mpl
import math
font = {'family' : 'DejaVu Sans',
        'weight' : 'normal',
        'size'   : 12}
mpl.rcParams['figure.figsize'] = (10,10)
mpl.rc('font', **font)
from tqdm import tqdm
from sklearn.preprocessing import StandardScaler
import datetime as dt
import joblib
from sklearn import linear_model
from sklearn.svm import LinearSVR
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.pipeline import make_pipeline
from sklearn.metrics import median_absolute_error, r2_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from scipy.optimize import curve_fit
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import cross_val_score
from datetime import datetime
```

```python
def line(x, a, b):
    #Basic straight line equation
	return a * x + b
```

```python
def second_degree(x, a, b, c):
    return a*x**2 + b*x + c
```

```python
def log_x(x, a, b):
    return a + b * np.log(x)
```

```python
def exp_x(x, a, b):
    return x**a + b
```

```python
def get_start_end_unix(df, start, end):
    return df.loc[(pd.Timestamp(start) - pd.Timestamp("1970-01-01")) // pd.Timedelta('1s'):(pd.Timestamp(end) - pd.Timestamp("1970-01-01")) // pd.Timedelta('1s')]
```

```python
def get_ttf(df, target):
    popt_max, _ = curve_fit(line, (df.index - pd.Timestamp("1970-01-01")) // pd.Timedelta('1s'),df['Inlet Filter Differential Pressure (bar)'])
    # summarize the parameter values
    a_max, b_max = popt_max
    y_line = line((df.index - pd.Timestamp("1970-01-01")) // pd.Timedelta('1s'), a_max, b_max)
    plt.plot((df.index - pd.Timestamp("1970-01-01")) // pd.Timedelta('1s'), y_line, '--', color = 'k')
    plt.scatter((df.index - pd.Timestamp("1970-01-01")) // pd.Timedelta('1s'),df['Inlet Filter Differential Pressure (bar)'])
    print(datetime.utcfromtimestamp(int((target-b_max)/a_max)).strftime('%Y-%m-%d %H:%M:%S'))
    plt.scatter(int((target-b_max)/a_max),target)
```

```python
def get_ttf_2nd(df, target):
    popt_max, _ = curve_fit(second_degree, (df.index - pd.Timestamp("1970-01-01")) // pd.Timedelta('1s'),df['Inlet Filter Differential Pressure (bar)'],  maxfev=5000)
    # summarize the parameter values
    a_max, b_max, c_max = popt_max
    y_line = second_degree((df.index - pd.Timestamp("1970-01-01")) // pd.Timedelta('1s'), a_max, b_max, c_max)
    plt.plot((df.index - pd.Timestamp("1970-01-01")) // pd.Timedelta('1s'), y_line, '--', color = 'k')
    plt.scatter((df.index - pd.Timestamp("1970-01-01")) // pd.Timedelta('1s'),df['Inlet Filter Differential Pressure (bar)'])
    print(datetime.utcfromtimestamp(int((math.sqrt((-4*a_max*c_max)+(4*a_max*target)+(b_max**2))-b_max)/(2*a_max))).strftime('%Y-%m-%d %H:%M:%S'))
    plt.scatter(int((math.sqrt((-4*a_max*c_max)+(4*a_max*target)+(b_max**2))-b_max)/(2*a_max)),target)
```

```python
def get_ttf_log(df, target):
    popt_max, _ = curve_fit(log_x, (df.index - pd.Timestamp("1970-01-01")) // pd.Timedelta('1s'),df['Inlet Filter Differential Pressure (bar)'])
    # summarize the parameter values
    a_max = popt_max
    y_line = exp_x((df.index - pd.Timestamp("1970-01-01")) // pd.Timedelta('1s'), a_max, b_max)
    plt.plot((df.index - pd.Timestamp("1970-01-01")) // pd.Timedelta('1s'), y_line, '--', color = 'k')
    plt.scatter((df.index - pd.Timestamp("1970-01-01")) // pd.Timedelta('1s'),df['Inlet Filter Differential Pressure (bar)'])
    print(datetime.utcfromtimestamp(int(np.exp((target/b_max)-(a_max/b_max)))).strftime('%Y-%m-%d %H:%M:%S'))
```

```python
def get_ttf_exp(df, target):
    popt_max, _ = curve_fit(exp_x, (df.index - pd.Timestamp("1970-01-01")) // pd.Timedelta('1s'),df['Inlet Filter Differential Pressure (bar)'],maxfev=5000)
    # summarize the parameter values
    a_max, b_max = popt_max
    y_line = exp_x((df.index - pd.Timestamp("1970-01-01")) // pd.Timedelta('1s'), a_max, b_max)
    plt.plot((df.index - pd.Timestamp("1970-01-01")) // pd.Timedelta('1s'), y_line, '--', color = 'k')
    plt.scatter((df.index - pd.Timestamp("1970-01-01")) // pd.Timedelta('1s'),df['Inlet Filter Differential Pressure (bar)'])
    #print(datetime.utcfromtimestamp(int(np.log(target/a_max)/np.log(b_max))).strftime('%Y-%m-%d %H:%M:%S'))
```

```python
def get_ttf_VEAS(df, target):
    popt_max, _ = curve_fit(line, (df.index - pd.Timestamp("1970-01-01")) // pd.Timedelta('1s'),df['Inlet Filter dp (Pa)'])
    # summarize the parameter values
    a_max, b_max = popt_max
    y_line = line((df.index - pd.Timestamp("1970-01-01")) // pd.Timedelta('1s'), a_max, b_max)
    plt.plot((df.index - pd.Timestamp("1970-01-01")) // pd.Timedelta('1s'), y_line, '--', color = 'k')
    plt.scatter((df.index - pd.Timestamp("1970-01-01")) // pd.Timedelta('1s'),df['Inlet Filter dp (Pa)'])
    print(datetime.utcfromtimestamp(int((target-b_max)/a_max)).strftime('%Y-%m-%d %H:%M:%S'))
```
