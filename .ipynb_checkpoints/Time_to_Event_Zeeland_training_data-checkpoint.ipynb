```python
import pandas as pd
import seaborn as sns
import numpy as np
import glob
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib as mpl
import math
font = {'family' : 'DejaVu Sans',
        'weight' : 'normal',
        'size'   : 12}
mpl.rcParams['figure.figsize'] = (10,10)
mpl.rc('font', **font)
from tqdm import tqdm
from sklearn.preprocessing import StandardScaler
import datetime as dt
import joblib
from sklearn import linear_model
from sklearn.svm import LinearSVR
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.pipeline import make_pipeline
from sklearn.metrics import median_absolute_error, r2_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from scipy.optimize import curve_fit
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import cross_val_score
```

```python
files = glob.glob('Export_16-09-2021-143815/Blower 305C01A/LAST/*.csv')
```

```python
df = pd.DataFrame()
for x in files:
    temp_df = pd.read_csv(x)
    temp_df['timestamp'] = pd.to_datetime(temp_df['timestamp'], format='%Y-%m-%d', errors='ignore')
    temp_df = temp_df.set_index('timestamp')
    df = pd.concat([df, temp_df], axis=1)
#df = df.drop(columns = 'Unnamed: 0')
```

```python
sns.heatmap(df)
```

```python
sns.scatterplot(data = df, y = 'Inlet Filter Differential Pressure (bar)', x='timestamp')
```

```python
df.to_csv('Zeeland_airfilter_val_filter.csv')
```

```python
from sklearn.cluster import KMeans
```

```python
X = df.ffill().dropna()
```

```python
X=X.loc['2021-4-15':'2021-8-12']
```

```python
kmeans = KMeans(n_clusters=4, random_state=0).fit(X)
```

```python
X['labels']=kmeans.labels_
```

```python
sns.scatterplot(data = X, x='timestamp', y= 'MDM current (A)', hue = 'labels')
```

```python
sns.scatterplot(data = X, x='timestamp', y= 'Inlet Filter Differential Pressure (bar)', hue = 'labels')
```

```python
sns.scatterplot(data = X, x='MDM current (A)', y= 'Inlet Filter Differential Pressure (bar)', hue='labels')
```

```python
sns.scatterplot(data = X, x='Outlet Flow (kg/s)', y= 'Inlet Filter Differential Pressure (bar)', hue='labels')
```
